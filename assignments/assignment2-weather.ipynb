# Weather Analysis

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
# Loading the CSV data

url = "https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv"
data = pd.read_csv(url)
```

## Examine Data

```python
# Displaing the first few rows of the data

data.head()
```

```python
# Convert the time column to datetime format if necessary

# Replaceing the 'datetime' with the correct column name if it differs

data['reportStartDateTime'] = pd.to_datetime(data['reportStartDateTime'])
```

```python
# Create a plot of dryBulbTemperature_Celsius over time
plt.figure(figsize=(12, 6))
plt.plot(data['reportStartDateTime'], data['dryBulbTemperature_Celsius'], color='red', linewidth=1.5)

# Customize the plot with title and labels
plt.title('Temperature Over Time', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)

# Adding grid lines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
